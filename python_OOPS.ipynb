```python
#define class student..
class student:
  name="Krishna"
  place="kollam"
  def displayName(self):
    print(self.name)
    print(self.place)
s=student()
s.displayName()
```

```python
#define class employee..
class employee:
  id="1"
  name="krishna"
  salary=25000
  def displayDetails(self):
    print("employee id=",self.id)
    print("employee name=",self.name)
    print("employee salary",self.salary)
e=employee()
e.displayDetails()
```

```python
#to practice constructor usage
class person:
  def __init__(self,name):
    self.name=name
  def display(self):
    print("name=",self.name)
p=person("krishna")
p.display()
```

```python
class person:
  def __init__(self,name,age):
    self.name=name
    self.age=age
  def display(self):
    print("name=",self.name)
    print("age=",self.age)
p=person("krishna","22")
p.display()
```

```python
#to create a new class student with a constructor to intialize the variables student id ,name,dep,and roll num
class student:
  def __init__(self,id,name,dep,roll):
    self.id=id
    self.name=name
    self.dep=dep
    self.roll=roll
  def display(self):
    print("student id=",self.id)
    print("student name=",self.name)
    print("student dep=",self.dep)
    print("student roll=",self.roll)
s=student("1","krishna","MCA","38")
s.display()
```
