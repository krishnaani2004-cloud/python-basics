```python
students = []


def add_student():
    roll = int(input("Enter Roll Number: "))
    name = input("Enter Name: ")
    python_marks = int(input("Enter Python Marks: "))
    ds_marks = int(input("Enter Data Science Marks: "))
    maths_marks = int(input("Enter Mathematics Marks: "))

    student = {
        "roll": roll,
        "name": name,
        "python": python_marks,
        "datascience": ds_marks,
        "maths": maths_marks
    }

    students.append(student)
    print("Student Added Successfully!\n")

add_student()

for student in students:
    print(f"Roll Number: {student['roll']}")
    print(f"Name: {student['name']}")
    print("Student displayed")


def search_student():
    roll = int(input("Enter Roll Number to Search: "))

    for s in students:
        if s["roll"] == roll:
            print("Student Found:")
            print(s)
            return
search_student()

#to find the topper
def find_topper():
    topper = None
    highest_marks = 0
    for student in students:
        total_marks = student["python"] + student["datascience"] + student["maths"]
        if total_marks > highest_marks:
            highest_marks = total_marks
            topper = student
    return topper
topper = find_topper()
print("Topper:")
print(topper)





```
